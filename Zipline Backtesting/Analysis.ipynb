### YFinance Data Results

```python
results = pd.read_pickle("macd_yfinance_results.pickle")
```

```python
results
```

```python
import matplotlib.pyplot as plt

plt.rcParams["axes.formatter.useoffset"] = False

results.portfolio_value.plot()
```

```python
results.columns
```

```python
print("Final Portfolio Value:")
print(results["portfolio_value"].iloc[-1])

# 1.16 for example would mean +116.99%
print("\nTotal Return:")
print(results["algorithm_period_return"].iloc[-1])

# Max drawdown of -0.2441 for example means the worst peak to through loss was -24.41%
print("\nMax Drawdown:")
print(results["max_drawdown"].iloc[-1])

# Sharpe ratio of 0.6863 for example means weak to moderate rewards considering the risk (1-2 is good, >2 or >3 is great, <0 is very bad)
print("\nSharpe Ratio:")
print(results["sharpe"].iloc[-1])

print("\nTotal Orders:")
print(results["orders"].apply(len).sum())

print("\nTotal Transactions:")
print(results["transactions"].apply(len).sum())

print("\nLast Positions:")
print(results["positions"].iloc[-1])
```

### Multi Result

```python
results = pd.read_pickle("multi_stock_macd_yfinance_results.pickle")
```

```python
results
```

```python
import matplotlib.pyplot as plt

plt.rcParams["axes.formatter.useoffset"] = False

results.portfolio_value.plot()
```

```python
results.columns
```

```python
print("Final Portfolio Value:")
print(results["portfolio_value"].iloc[-1])

# 1.16 for example would mean +116.99%
print("\nTotal Return:")
print(results["algorithm_period_return"].iloc[-1])

# Max drawdown of -0.2441 for example means the worst peak to through loss was -24.41%
print("\nMax Drawdown:")
print(results["max_drawdown"].iloc[-1])

# Sharpe ratio of 0.6863 for example means weak to moderate rewards considering the risk (1-2 is good, >2 or >3 is great, <0 is very bad)
print("\nSharpe Ratio:")
print(results["sharpe"].iloc[-1])

print("\nTotal Orders:")
print(results["orders"].apply(len).sum())

print("\nTotal Transactions:")
print(results["transactions"].apply(len).sum())

print("\nLast Positions:")
print(results["positions"].iloc[-1])
```

```python
import matplotlib.pyplot as plt

last_positions = results.positions.iloc[-1]
last_cash = results.ending_cash.iloc[-1]

labels = []
values = []

for position in last_positions:
    symbol = position["sid"].symbol
    value = position["amount"] * position["last_sale_price"]

    labels.append(f"{symbol}\n${value:,.2f}")
    values.append(value)

if last_cash > 0:
    labels.append(f"Cash\n${last_cash:,.2f}")
    values.append(last_cash)

plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Final Portfolio Allocation")
plt.show()
```
