### Performance Increase Classification (CTGAN)

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sdv.metadata import SingleTableMetadata
from sdv.single_table import CTGANSynthesizer
```

```python
df = load_breast_cancer(as_frame=True).frame

train, test = train_test_split(df, test_size=0.2)

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(train)
metadata.update_column("target", sdtype="categorical")

synth = CTGANSynthesizer(metadata, epochs=100)
synth.fit(train)

synthetic = synth.sample(num_rows=len(train))

X_train = train.drop(columns="target")
y_train = train["target"]

augmented = pd.concat([train, synthetic], ignore_index=True)
X_aug = augmented.drop(columns="target")
y_aug = augmented["target"]

X_test = test.drop(columns="target")
y_test = test["target"]

baseline = RandomForestClassifier()
baseline.fit(X_train, y_train)

augmented_model = RandomForestClassifier()
augmented_model.fit(X_aug, y_aug)

print(baseline.score(X_test, y_test))
print(augmented_model.score(X_test, y_test))
```

### Balancing Classes
#### TVAE (Filtering)

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sdv.metadata import SingleTableMetadata
from sdv.single_table import TVAESynthesizer

df = load_breast_cancer(as_frame=True).frame

train, test = train_test_split(df, test_size=0.2)

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(train)
metadata.update_column("target", sdtype="categorical")

synth = TVAESynthesizer(metadata, epochs=100)
synth.fit(train)

synthetic = synth.sample(len(train)*5)

minority = train.target.value_counts().idxmin()
needed = train.target.value_counts().max() - train.target.value_counts().min()

balanced = pd.concat([train, synthetic[synthetic.target == minority].head(needed)])

print(train.target.value_counts())
print(balanced.target.value_counts())

X_train = train.drop(columns="target")
y_train = train["target"]

X_balanced = balanced.drop(columns="target")
y_balanced = balanced["target"]

X_test = test.drop(columns="target")
y_test = test["target"]

model = RandomForestClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))

balanced_model = RandomForestClassifier()
balanced_model.fit(X_balanced, y_balanced)
print(balanced_model.score(X_test, y_test))
```

#### CTGAN (Conditional Generation)

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sdv.metadata import SingleTableMetadata
from sdv.single_table import CTGANSynthesizer
from sdv.sampling import Condition

df = load_breast_cancer(as_frame=True).frame

train, test = train_test_split(df, test_size=0.2)

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(train)
metadata.update_column("target", sdtype="categorical")

synth = CTGANSynthesizer(metadata, epochs=100)
synth.fit(train)

minority = train.target.value_counts().idxmin()
needed = train.target.value_counts().max() - train.target.value_counts().min()

condition = Condition(num_rows=int(needed), column_values={"target": minority})

synthetic = synth.sample_from_conditions([condition])

balanced = pd.concat([train, synthetic])

print(train.target.value_counts())
print(balanced.target.value_counts())

X_train = train.drop(columns="target")
y_train = train["target"]

X_balanced = balanced.drop(columns="target")
y_balanced = balanced["target"]

X_test = test.drop(columns="target")
y_test = test["target"]

model = RandomForestClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))

balanced_model = RandomForestClassifier()
balanced_model.fit(X_balanced, y_balanced)
print(balanced_model.score(X_test, y_test))
```

### Rare-Case Training

```python
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

from sdv.metadata import SingleTableMetadata
from sdv.single_table import TVAESynthesizer

df = fetch_california_housing(as_frame=True).frame
train, test = train_test_split(df, test_size=0.2)

rare = train[train.MedHouseVal > 4.5]

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(rare)

synth = TVAESynthesizer(metadata, epochs=100)
synth.fit(rare)

rare_synth = synth.sample(1000)
augmented = pd.concat([train, rare_synth], ignore_index=True)

X_train = train.drop(columns="MedHouseVal")
y_train = train["MedHouseVal"]

X_augmented = augmented.drop(columns="MedHouseVal")
y_augmented = augmented["MedHouseVal"]

X_test = test.drop(columns="MedHouseVal")
y_test = test["MedHouseVal"]

model = RandomForestRegressor()
model.fit(X_train, y_train)
preds = model.predict(X_test)
print(mean_absolute_error(y_test, model.predict(X_test)))

augmented_model = RandomForestRegressor()
augmented_model.fit(X_augmented, y_augmented)
print(mean_absolute_error(y_test, augmented_model.predict(X_test)))
```

```python
rare_test = test[test.MedHouseVal > 4.5]

X_rare_test = rare_test.drop(columns="MedHouseVal")
y_rare_test = rare_test["MedHouseVal"]

model = RandomForestRegressor()
model.fit(X_train, y_train)
print(mean_absolute_error(y_rare_test, model.predict(X_rare_test)))

augmented_model = RandomForestRegressor()
augmented_model.fit(X_augmented, y_augmented)
print(mean_absolute_error(y_rare_test, augmented_model.predict(X_rare_test)))
```
