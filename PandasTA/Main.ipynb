### Technical Analysis Indicators

Full List: https://xgboosted.github.io/pandas-ta-classic/indicators.html

```python
data['sma_50'] = data.ta.sma(length=50)
data['ema_50'] = data.ta.sma(length=50)
data["rsi_14"]  = data.ta.rsi(length=14)

data = data.join(data.ta.bbands(length=20))
data = data.join(data.ta.macd(fast=8, slow=21))
```

### Strategies

```python
NeuralNineStrategy = ta.Strategy(
  name="Momentum and Stuff",
  ta=[
      {"kind":"sma", "length":50},
      {"kind":"ema", "length":50},
      {"kind":"rsi", "length":14},
      {"kind":"bbands", "length":20},
      {"kind":"macd", "fast": 8, "slow": 21}
  ]
)

data = yf.Ticker('AAPL').history(period='5y')
data.ta.strategy(NeuralNineStrategy)
```

```python
data.iloc[50:]
```

### Create Trading Signals & Visualize Them Easily

#### EMA Crossing Strategy

```python
data = yf.Ticker('AAPL').history(period='5y')

data["ema20"] = ta.ema(data["Close"], length=20)
data["ema50"] = ta.ema(data["Close"], length=50)
data["ema20_x_ema50"] = ta.cross(data["ema20"], data["ema50"])
```

```python
data
```

```python
data.ema20_x_ema50.value_counts()
```

#### RSI Threshold Strategy

```python
# This Example Needs Pandas <3.0, otherwise a deprecated method used by pandas-ta-classic is removed already
# In older versions -> Only deprecation warning
data = yf.Ticker('AAPL').history(period='5y')

data["rsi14"] = ta.rsi(data["Close"], length=14)
data = data.join(ta.xsignals(data["rsi14"], xa=30, xb=70))
```

```python
data
```

```python
import matplotlib.pyplot as plt

entries = data.index[data["TS_Entries"] == 1]
exits = data.index[data["TS_Exits"] == 1]

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 7), gridspec_kw={"height_ratios": [3, 1]})

ax1.plot(data.index, data["Close"], label="Price")
ax1.scatter(entries, data.loc[entries, "Close"], marker="^", s=70, label="Buy", color='green')
ax1.scatter(exits,   data.loc[exits,   "Close"], marker="v", s=70, label="Sell", color='red')
ax1.set_title("AAPL — Trading Signals")
ax1.grid(True, alpha=0.25)
ax1.legend(loc="upper left", frameon=False)

ax2.axhline(30, linestyle="--", linewidth=1, color='gray')
ax2.axhline(70, linestyle="--", linewidth=1, color='gray')
ax2.plot(data.index, data["rsi14"])
ax2.set_ylim(0, 100)
ax2.grid(True, alpha=0.25)
ax2.set_title('RSI (14)')

plt.tight_layout()
plt.show()
```
